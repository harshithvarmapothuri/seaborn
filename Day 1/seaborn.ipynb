```python
import seaborn as sns
```

# Load iris Dataset

```python
iris=sns.load_dataset("iris")
```

```python
iris
```

# Scatterplot

```python
sns.scatterplot(x=iris.sepal_length,y=iris.sepal_width)
```

```python
sns.scatterplot(x=iris.sepal_length,y=iris.petal_length)
```

```python
sns.scatterplot(x=iris.sepal_length,y=iris.petal_width)
```

# Load Tips Dataset

```python
tips=sns.load_dataset("tips")
```

```python
tips
```

# Relplot

```python
sns.relplot(x=tips.total_bill,y=tips.tip,data=tips,hue="size")
```

```python
sns.relplot(x=tips.total_bill,y=tips.tip,data=tips,style="size")
```

```python
sns.relplot(x=tips.total_bill,y=tips.tip,data=tips,hue="sex",style="size")
```

# Catplot

```python
sns.catplot(x=tips.sex,y=tips.total_bill)
```

# Jointplot

```python
sns.jointplot(x=tips.total_bill,y=tips.tip)
```

# Displot

```python
sns.displot(x=tips["size"])
```
